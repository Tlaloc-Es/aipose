```python
import warnings
warnings.filterwarnings('ignore')
import cv2
import numpy as np
from numpy import ndarray
from typing import List
from aipose.models.yolov7 import YoloV7PoseKeypoints, YoloV7Pose
from aipose.plot import plot
import matplotlib.pyplot as plt
```

```python
model = YoloV7Pose()
```

```python
%matplotlib inline 
image: ndarray = cv2.imread('../person.jpg')
prediction: List[YoloV7PoseKeypoints]
image_tensor: ndarray
prediction, image_tensor = model(image)
img = plot(image_tensor, np.array([value.get_raw_keypoint() for value in prediction]), plot_image=True, return_img=True)
plt.show()
```
